```python
import pandas as pd

df = pd.read_csv(r"C:\Users\annie\OneDrive\Documents\Projects\ocean_clean_project\data\raw\coastal_cleanup_US_detailed.csv")
df.head()
```

```python
# Day 2: Data Cleaning – Global Coastal Cleanup Project
# it give number of rows and columns are in file 
df.shape
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace(r"[^\w]", "", regex=True)
)

df.columns
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['cleanup_date'] = pd.to_datetime(df['cleanup_date'], errors='coerce')
df['cleanup_year'] = df['cleanup_date'].dt.year

df[['cleanup_date', 'cleanup_year']].head()
```

```python
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(0)
```

```python
df.isnull().sum().sum()
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
df.to_csv(r"C:\Users\annie\OneDrive\Documents\Projects\ocean_clean_project\data\raw\coastal_cleanup_US_cleaned.csv", index=False)
```
